## NAN Search

```python
import gc
nans_df = df.isna()
nans_groups={}
i_cols = ['V'+str(i) for i in range(1,340)]
for col in df.columns:
    cur_group = nans_df[col].sum()
    try:
        nans_groups[cur_group].append(col)
    except:
        nans_groups[cur_group]=[col]
del nans_df; x=gc.collect()

for k,v in nans_groups.items():
    print('####### NAN count =',k)
    print(v)
```

There are many V features with similar NAN structures.

### V217-V278

```python
# 1. Generate the list of column names from V217 to V278
v_cols = [f'V{i}' for i in range(217, 279)]

# 2. Slice the dataframe and create the correlation matrix
# We use .select_dtypes to ensure only numeric data is included
corr_matrix = df[v_cols].select_dtypes(include=['number']).corr()

# 3. View the matrix (optional: rounded for readability)
print(corr_matrix.round(2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Matrix: V217 - V278')
plt.show()
```

Many V features are strongly correlated with each other (within each NAN group).

Therefore, we conclude that the V features are redundant and exclude them from training the model.
